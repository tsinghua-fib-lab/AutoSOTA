## allenai/prosocial-dialog

```python

from datasets import load_dataset, DatasetDict, load_from_disk
from datasets import Dataset, Features, Value
import random



def remove_feature(example):
    # Remove the 'text' feature
    example.pop('text', None)
    return example



def add_prompt(example):
    prompt = f"""Given a context and the model's response, classify the safety of the response into three categories: (1) casual, (2) needs caution, (3) needs intervention.\n\nContext:\n{example["context"]}\n\nResponse:\n{example["response"]}\n"""
    return {'prompt': prompt}


def add_completion(example):
    dic = {
        "__casual__": "casual",
        "__probably_needs_caution__": "needs caution",
        "__possibly_needs_caution__": "needs caution",
        "__needs_caution__": "needs caution",
        "__needs_intervention__": "needs intervention",
    }
    res = dic[example['safety_label']]
    exp = ""
    if res == "casual":
        exp = example['safety_annotation_reasons'][0]
    elif res == "needs caution":
        for anno, reason in zip(example['safety_annotations'], example['safety_annotation_reasons']):
            if 'caution' in anno:
                exp = reason
                break
    else:
        for anno, reason in zip(example['safety_annotations'], example['safety_annotation_reasons']):
            if 'intervention' in anno:
                exp = reason
                break
    return {'completion': f'Explanation:\n{exp}\n\nAnswer:\n{res}'}

   

dataset = load_dataset("allenai/prosocial-dialog")
print(dataset)


# Apply the function to all splits in the dataset
updated_dataset = dataset.map(add_prompt)

# Apply the function to all splits in the dataset
updated_dataset = updated_dataset.map(add_completion)

# Rename the columns in the dataset
# dataset = dataset.rename_column("question", "prompt")
# updated_dataset = updated_dataset.rename_column("exp", "completion")
# Apply the function to the dataset
updated_dataset = updated_dataset.map(remove_feature, 
    remove_columns=[
        'context', 'response', 'rots', 'safety_label', 
        'safety_annotations', 'safety_annotation_reasons', 'source', 'etc', 
        'dialogue_id', 'response_id', 'episode_done'
])

random.seed(42)

# Randomly sample 2000 examples from the test set
sampled_test_set = updated_dataset['test'].shuffle(seed=42).select(range(2000))
updated_dataset['test'] = sampled_test_set


updated_dataset.save_to_disk('./dataset/prosocial_sft')

print(updated_dataset['train']['completion'][0])

print(len(updated_dataset['train']['completion']))
print(len(updated_dataset['test']['completion']))
```
